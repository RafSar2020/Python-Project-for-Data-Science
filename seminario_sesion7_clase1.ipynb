```python
def precsv(filename):
    with open(path+filename,encoding="ISO-8859-1") as infile:
        content = infile.read()
        soup = BeautifulSoup(content, 'html.parser')
        table = soup.select_one("table")
        return [[cell.text.replace("\n","").replace("\xa0"," ").replace("\u200b", "").replace("\t", "").replace("-\n", "").replace("\u200b\n", "") for cell in row.find_all(["td","th"])] for row in table.select("tr")]
```
