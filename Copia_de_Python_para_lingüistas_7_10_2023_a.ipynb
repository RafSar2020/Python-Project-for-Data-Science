```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el polinomio de quinto grado
# Por ejemplo, y = 2x^5 - 3x^4 + 4x^3 - 5x^2 + 6x - 7
coefficients = [2, -3, 4, -5, 6, -7]

# Generar valores de x
x = np.linspace(-10, 10, 400)  # Rango de valores de x

# Calcular los valores de y para cada valor de x
y = np.polyval(coefficients, x)

# Crear el gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de Quinto Grado')
plt.grid(True)
plt.show()
```

Dado el número ordinal de un mes escribe el nombre del mes y el número dedías que tiene suponiendo que el año no es bisiesto

```python
# Paso 1: Entrada de datos
numero_mes = int(input("Introduce el número ordinal del mes (1-12): "))

# Paso 2: Conversión de número a nombre de mes
meses = {
    1: "enero",
    2: "febrero",
    3: "marzo",
    4: "abril",
    5: "mayo",
    6: "junio",
    7: "julio",
    8: "agosto",
    9: "septiembre",
    10: "octubre",
    11: "noviembre",
    12: "diciembre"
}

# Paso 3: Asociación de número de días
dias_mes = {
    1: 31,
    2: 28,
    3: 31,
    4: 30,
    5: 31,
    6: 30,
    7: 31,
    8: 31,
    9: 30,
    10: 31,
    11: 30,
    12: 31
}

# Paso 4: Manejo de entradas inválidas
if numero_mes < 1 or numero_mes > 12:
    print("Número de mes inválido. Debe estar en el rango de 1 a 12.")
else:
    # Paso 5: Salida de resultados
    nombre_mes = meses[numero_mes]
    num_dias = dias_mes[numero_mes]
    print(f"El mes {nombre_mes} tiene {num_dias} días.")
```

```python
#Otra forma de hacerlo con listas
# Paso 1: Entrada de datos
numero_mes = int(input("Introduce el número ordinal del mes (1-12): "))

# Paso 2: Verificación de número de mes válido
if 1 <= numero_mes <= 12:
    # Paso 3: Asociación de número de días usando listas
    meses = ["enero", "febrero", "marzo", "abril", "mayo", "junio", "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre"]
    dias_por_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    # Paso 4: Obtención del nombre del mes y la cantidad de días
    nombre_mes = meses[numero_mes - 1]
    num_dias = dias_por_mes[numero_mes - 1]

    # Paso 5: Salida de resultados
    print(f"El mes {nombre_mes} tiene {num_dias} días.")
else:
    # Manejo de entrada inválida
    print("Número de mes inválido. Debe estar en el rango de 1 a 12.")
```

```python
#No has sabido hacerlo. Respuesta chatgpt
texto = "Estoy muy contento de estar aquí, porque es muy interesante"
contador_vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}

for caracter in texto:
    # Verifica si el carácter es una vocal
    if caracter.lower() in 'aeiou':
        # Actualiza el contador de la vocal correspondiente
        contador_vocales[caracter.lower()] += 1

# Imprime el recuento de cada vocal
for vocal, count in contador_vocales.items():
    print(f'Número de "{vocal}": {count}')
```

```python
#Sin diccionario
texto = "Estoy muy contento de estar aquí, porque es muy interesante"

count_a = 0
count_e = 0
count_i = 0
count_o = 0
count_u = 0

for caracter in texto:
    if caracter.lower() == 'a':
        count_a += 1
    elif caracter.lower() == 'e':
        count_e += 1
    elif caracter.lower() == 'i':
        count_i += 1
    elif caracter.lower() == 'o':
        count_o += 1
    elif caracter.lower() == 'u':
        count_u += 1

# Imprime el recuento de cada vocal
print(f'Número de "a": {count_a}')
print(f'Número de "e": {count_e}')
print(f'Número de "i": {count_i}')
print(f'Número de "o": {count_o}')
print(f'Número de "u": {count_u}')
```
